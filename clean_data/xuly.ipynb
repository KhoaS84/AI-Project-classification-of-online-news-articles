```python
import pandas as pd

# Đọc dữ liệu
df = pd.read_csv('/content/drive/MyDrive/FILE/Năm 3/Khoa học dữ liệu/Nhóm 09 - PTDL Báo điện tử/data/data_cleaned.csv')

# Tạo từ điển ánh xạ các giá trị sai lệch/lặp → chuẩn hóa
mapping = {
    'van-hoa': 'Văn hóa',
    'Văn hóa': 'Văn hóa',

    'giao-duc': 'Giáo dục',
    'Giáo dục': 'Giáo dục',

    'du-lich': 'Du lịch',
    'Du lịch': 'Du lịch',

    'kinh-doanh': 'Kinh doanh',
    'Kinh doanh': 'Kinh doanh',

    'the-gioi': 'Thế giới',
    'Thế giới': 'Thế giới',

    'giai-tri': 'Giải trí',
    'Giải trí': 'Giải trí',

    'thoi-su': 'Thời sự',
    'Thời sự': 'Thời sự',

    'the-thao': 'Thể thao',
    'Thể thao': 'Thể thao',

    'suc-khoe': 'Sức khỏe',
    'Sức khỏe': 'Sức khỏe',

    'khoa-hoc': 'Khoa học',
    'Khoa học': 'Khoa học',
}

# Chuẩn hóa cột category
df['category'] = df['category'].map(mapping)

# Xem các giá trị không khớp (nếu còn)
unmatched = df[~df['category'].isin(mapping.values())]['category'].unique()
print("Các giá trị chưa được ánh xạ:", unmatched)
```

```python
# Ghi đè dữ liệu đã xử lý vào file cũ
df.to_csv('/content/drive/MyDrive/FILE/Năm 3/Khoa học dữ liệu/Nhóm 09 - PTDL Báo điện tử/data/data_cleaned.csv', index=False)
```

```python
# Lấy tối đa 600 mẫu cho mỗi category
sampled_df = df.groupby('category', group_keys=False).apply(lambda x: x.sample(n=min(len(x), 600), random_state=42))

# Kiểm tra kết quả
print(sampled_df['category'].value_counts())

# Ghi ra file mới nếu cần
sampled_df.to_csv('/content/drive/MyDrive/FILE/Năm 3/Khoa học dữ liệu/Nhóm 09 - PTDL Báo điện tử/data/data_sampled.csv', index=False)
```
